```python
# Encoding- convert catergorical data into numerical data

```

```python
df["school"]=df["school"].map({"GP":1,"MS":0})
```

```python
df["sex"]=df["sex"].map({"M":1,"F":0})
```

```python
df["address"]=df["address"].map({"U":1,"R":0})
```

```python
df["famsize"]=df["famsize"].map({"GT3":1,"LE3":0})
```

```python
df["Pstatus"]=df["Pstatus"].map({"A":1,"T":0})
```

```python
# to convert catergorical data into numerical by using dummies:-

df=pd.get_dummies(df,drop_first=True)
```

```python
# separate x and y

x=df.drop(["G3"],axis=1)
y=df["G3"]
```

```python
# Train - Test - Split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Scaling

from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)

```

```python
x_test=scaler.transform(x_test)
```

```python
 # model_train

from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(n_estimators=100,random_state=42,max_depth=10,min_samples_split=30,min_samples_leaf=2)

model
```

```python
model.fit(x_train,y_train)
```

```python
# model_prediction

y_pred=model.predict(x_test)
```

```python
# model_accuracy:-

from sklearn.metrics import r2_score,accuracy_score

print("Accuracy of the model =",r2_score(y_test,y_pred))
```
